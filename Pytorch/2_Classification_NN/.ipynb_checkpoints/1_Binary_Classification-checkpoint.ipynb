```python
german = pd.read_csv("german.csv",index_col= 0) # index_col=0 해두면 Unnamed: 0 사라짐
print(german.shape) ## 1000개의 데이터 20개의 feature 
```

```python
german.iloc[0,:] ## Qualitative data가 많아서 다 dummy화 시켜준다.
```

```python
german.isnull().sum() ## 결측치 확인
```

```python
german_categorized = pd.get_dummies(german)
german_categorized.head()
```

```python
## Testset을 100개 남겨놓고 나머지 900개의 dataset을 train valid set으로 남겨둡니다. 

data_X = german_categorized.drop(['y'], axis=1)
data_y = german_categorized['y']

test_X = data_X.iloc[900:,:]
test_y = data_y.iloc[900:]

train_data_X = data_X.iloc[:900,:]
train_data_y = data_y.iloc[:900]
```

```python
train_X,valid_X,train_y,valid_y = train_test_split(train_data_X,train_data_y,test_size=0.15,random_state=0)
```

```python
# train, valid dataset을 출력할 수 있는 class를 만듭니다.
class MakeData(Dataset):
    
    def __init__(self, X_data, y_data):
        self.X_data = X_data
        self.y_data = y_data
        
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
        
    def __len__ (self):
        return len(self.X_data)
```

```python
train_X.shape
```

```python
# normalize까지만 전처리를 해보겠습니다.
train_data = MakeData(np.array(normalize(train_X)), np.array(train_y))
test_data = MakeData(np.array(normalize(valid_X)), np.array(valid_y))
```

```python
train_data.__getitem__(0)
```

## 3. Model & Optimizer

```python
class Binary_Classification(nn.Module):
    def __init__(self):
        super(Binary_Classification, self).__init__()
       
        self.Layer_1 = nn.Linear(61, 30) # Layer를 2개 쌓았다.
        self.Layer_2 = nn.Linear(30, 6)

        
        self.dropout = nn.Dropout(0.5)
        self.sigmoid = nn.Sigmoid()
        self.out = nn.Linear(6,1)
        
        
    def forward(self, inputs):
        x = self.Layer_1(inputs)
        x = self.sigmoid(x)
        x = self.dropout(x)
        
        x = self.Layer_2(x)
        x = self.sigmoid(x)
        x = self.dropout(x)
        
        
        x = self.out(x)
        x = self.sigmoid(x)
        return x 
        # cutoff 가 0.5이면 round(x)를 반환하면 됩니다
    
    def predict(self,test_inputs):
        # test에서는 dropout하면 안됩니다. 
        
        x = self.Layer_1(test_inputs)
        x = self.sigmoid(x)
        
        x = self.Layer_2(x)
        x = self.sigmoid(x)
        
        x= self.out(x)
        x = self.sigmoid(x)
        
        return x
```

```python
EPOCHS = 5000
BATCH_SIZE = 64

model = Binary_Classification()

# Loss 는 binary cross entropy
criterion = nn.BCELoss()

train_loader = DataLoader(dataset = train_data, batch_size=BATCH_SIZE, shuffle = True)
optimizer = optim.Adam(model.parameters(),lr = 0.01)
for epoch in range(EPOCHS):
    for X_batch, y_batch in train_loader:
        inputs = Variable(torch.Tensor(X_batch.float()))
        targets = Variable(torch.Tensor(y_batch.float())).unsqueeze(1)
        
        model.zero_grad()
        
        y_pred = model(inputs)
        
        loss = criterion(y_pred, targets)
        loss.backward()
        
        optimizer.step()
        
    if epoch % 1000 ==0:
        print(loss)
```

## 4. Check Trained Parameters
- validset이용해서 적절한 cutoff찾는다
- 이를 다시 testset에 적용

```python
y_pred = model.predict(torch.Tensor(normalize(valid_X)))
```

```python
def Binary(x,cutoff):
    if x < cutoff:
        return 0
    else:
        return 1
```

```python
f1_score = []
for i in [0.6,0.61,0.62,0.63,0.64,0.65,0.66,0.67,0.68,0.69,0.7,0.71,0.72,0.73,0.74,0.75]:
    Binary_output = [Binary(x,i) for x in y_pred]
    Binary_output = np.array(Binary_output) 
    f1 = metrics.f1_score(valid_y, Binary_output)
    f1_score.append(f1)
    
```

```python
f1_score
```

```python
## cutoff = 0.63에서 f1_score가 0.8435으로 가장 높게 나왔다. 

y_pred = model.predict(torch.Tensor(normalize(test_X)))
y_pred = [Binary(x,0.63) for x in y_pred] # cutoff 0.63으로 설정
test_y = np.array(test_y)
f1 = metrics.f1_score(y_pred, test_y)
accuracy = metrics.accuracy_score(y_pred, test_y)
```
